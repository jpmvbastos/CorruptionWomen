```python
import pandas as pd 
import numpy as np
```

```python
# Import code equivalence across TSE and IBGE
ids = pd.read_csv('/Users/jpmvbastos/Documents/GitHub/CorruptionDonations/Data/municipios_brasileiros_tse.csv')
ids = ids[['codigo_tse','codigo_ibge','capital']]
ids.columns = ['tse_code','ibgecode','capital']

data = ids
audit = pd.read_stata('/Users/jpmvbastos/Documents/GitHub/CorruptionDonations/Data/Audit/audited.dta')
audit['year'] = audit['year'].astype(int)
audit

for year in audit.year.unique():
     audited = audit.loc[(audit['year']==year) & (audit['audit']==1), ['ibgecode','audit','lottery']]
     data = data.merge(audited, on='ibgecode', how='left').rename(columns={'audit':f'audited_{year}',
                                                                           'lottery':f'lottery_{year}'})

data.replace(np.nan, 0, inplace=True)

data.rename(columns={'ibgecode':'cod_munic'}, inplace=True)

data.drop_duplicates(subset=['cod_munic'], inplace=True)

display(data)
```
